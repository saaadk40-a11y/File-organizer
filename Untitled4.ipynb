```python
import os
import shutil
from datetime import datetime

CATEGORIES = {
    "Images": [".jpg", ".jpeg", ".png", ".gif"],
    "Documents": [".doc", ".docx", ".txt", ".xlsx", ".csv"],
    "PDFs": [".pdf"],
    "Videos": [".mp4", ".mkv", ".avi"],
    "Audio": [".mp3", ".wav"],
    "Archives": [".zip", ".rar"],
    "Code": [".py", ".js", ".html", ".css"],
}


class FileOrganizer:

    def __init__(self, folder_path):
        self.folder_path = folder_path
        self.moved_files = []
        self.category_counts = {}

    def get_category(self, extension):
        for category, extensions in CATEGORIES.items():
            if extension.lower() in extensions:
                return category
        return "Others"

    def organize(self):
        for file_name in os.listdir(self.folder_path):
            file_path = os.path.join(self.folder_path, file_name)

            if os.path.isdir(file_path) or file_name == "organizer_report.txt":
                continue

            extension = os.path.splitext(file_name)[1]
            category = self.get_category(extension)

            category_folder = os.path.join(self.folder_path, category)
            os.makedirs(category_folder, exist_ok=True)

            new_path = os.path.join(category_folder, file_name)
            shutil.move(file_path, new_path)

            self.moved_files.append((file_name, category))
            self.category_counts[category] = self.category_counts.get(category, 0) + 1

    def save_report(self):
        report_path = os.path.join(self.folder_path, "organizer_report.txt")
        with open(report_path, "w") as f:
            f.write("FILE ORGANIZER REPORT\n")
            f.write(f"Timestamp: {datetime.now()}\n")
            f.write(f"Total files processed: {len(self.moved_files)}\n\n")

            f.write("Files per category:\n")
            for category, count in self.category_counts.items():
                f.write(f"  {category}: {count}\n")

            f.write("\nMoved files:\n")
            for file_name, category in self.moved_files:
                f.write(f"  {file_name} -> {category}\n")

        print(f"Report saved to: {report_path}")


def main():
    folder_path = input("Enter folder path to organize: ").strip()

    if not os.path.isdir(folder_path):
        print("Error: folder does not exist.")
        return

    organizer = FileOrganizer(folder_path)
    organizer.organize()
    organizer.save_report()
    print(f"Done! {len(organizer.moved_files)} files organized.")


if __name__ == "__main__":
    main()
```
